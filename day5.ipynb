```python
class User1:
  def __init__(self,pen,pencil):
    print("constructor of user 1")
    self.pen = pen
    self.pencil = pencil
  def penGet(self):
    print("pen is working: ",self.pen)
  def pencilGet(self):
    print("pencil is working: ",self.pencil)

class User2(User1):
  def __init__(self,pen,pencil,earser):
    print("constructor of user 2")
    # self.pen = pen
    # self.pencil = pencil
    super().__init__(pen,pencil)
    self.earser = earser
# object
u = User2("blue","black","doms")
u.penGet()
```

```python
# import math
# math.pow(3,2)
# math.factorial(5)
class math:
  def __init__(self):
    pass
  def pow(self,a,b):
    print(a**b)
  def factorial(self,a):
    count = 1
    for i in range(1,a+1):
      count *= i
    print(count)


m = math() # import vala kaam
m.pow(4,2)
m.factorial(5)
```

```python
class nikhil:
  def __init__(self,bike):
    self.bike = bike

  def usebike(self):
    print("bike used by: ",self.bike)

class rajendra(nikhil):
  def __init__(self,name,bike):
    super().__init__(bike)
    self.name = name

  def usecar(self):
    print("car used by: ",self.name)

# object instantiation and method calls are outside the class definition
u = rajendra("rajendra jangid","tvs by rajendra")
u.usecar() # Corrected from u.usename()
u.usebike()
```

```python
# encapsulation
class Student:
  def __init__(self,name,roll_no):
    self.__data = name # name is attribute
    self.roll_no = roll_no # roll_no is attribute
  def getname(self,password=None):
    if password == "123456":
      print(self.__data)
    else:
      print("password not match")

s = Student("rajendra",123)
# print(s.__data) # no attribute
print(s.roll_no)
s.getname("1234567")

# example
```

```python
# class variable
class Student:
  count = 0 # count is class variable
  def __init__(self,name,roll_no):
    print("constructor is working")
    self.name = name
    self.roll_no = roll_no
    # count = 13 # local variable
    Student.count = 12 # if you update the value of count by using class name
    print(count)

  def getname(self):
    print(self.name)


s = Student("rajendra",123)
s.getname()
print(s.count)
```

```python
# example of class variable
class User:
  name= ""
  roll_no = ""
  def __init__(self,name,roll_no):
    User.name = name
    User.roll_no = roll_no
  def getname(self):
    print(User.name)

u = User("rajendra",123)
u.getname()
u.name
```

```python
#overriding
class User1:
  def __init__(self,pen,pencil):
    self.pen = pen
    self.pencil = pencil
  def getpen(self):
    print("user1: ",self.pen)

class User2(User1):
  def __init__(self,pen,pencil,earser):
    self.earser = earser
    super().__init__(pen,pencil)
  def getpen(self):
    print("user2: ",self.pen)

u = User2("blue","black","doms")
u.getpen()
```
