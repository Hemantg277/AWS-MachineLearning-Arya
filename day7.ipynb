3. ones (2,3)

```python
# zeros
import numpy as np
# arrary 1d
arr = np.zeros(5)+7
print(arr)

# 2d
arr1 = np.zeros((2,3))+7 # 2 row and 3 cols
print(arr1)
```

```python
# ones
import numpy as np
# array 1d
arr = np.ones(5)
print(arr)

# 2d
arr1 = np.ones((2,3))
print(arr1)
```

```python
# full
import numpy as np
# array 1d
arr = np.full((5),7)
print(arr)

# 2d
arr1 = np.full((2,3),7)
print(arr1)
```

```python
# numpy | list
import numpy as np
import time as t

# for list
start = t.time()
sum1 = 0
for i in range(1000000):
  sum1 += i*2
end = t.time()
print("for list",sum1,end-start)

# for numpy | array (array work every element | element wise)
start1 = t.time()
sum2 = 0
arr = np.arange(1000000)*2
sum2 = arr.sum()
end1 = t.time()
print("for array ",sum2,end1-start1)
```

**Array properties**

1. shape
2. dimension | direction
3. size

```python
# array opr on zeros | ones | full
import numpy as np
arr = np.zeros(5)
print(arr)
print(arr.shape) # 5
print(arr.ndim) # 1d
print(arr.size) # 5
print(arr.dtype) # float

# 2d
print("2d ones start")
arr1 = np.zeros((3,2))
print(arr1)
print(arr1.shape) # 3,2
print(arr1.ndim) # 2d
print(arr1.size) # 6
print(arr1.dtype) # float
```

**Array Reshaping**

1. reshape()
2. flatten() -> create copy | shallow copy
3. ravel() -> return original | deep copy
4. transpose -> .T -> reverse

```python
# 1d array
import numpy as np
arr = np.arange(6)
print(arr)

# 2d
arr1 = arr.reshape(2,3) # 2 rows and 3 cols
print(arr1)
```

```python
# flatten | copy
import numpy as np
# 2d
arr = np.arange(6)
arr1 = arr.reshape(2,3)
print(arr1)
# flatten
arr_new = arr1.flatten()
arr_new[0] = 100
print(arr1)
print(arr_new)
```

```python
# ravel | original
import numpy as np
# 2d
arr = np.arange(6).reshape(2,3)
print(arr)
# ravel
arr_new1 = arr.ravel()
arr_new1[0] = 200
print(arr)
print(arr_new1)
```

```python
# transpose
import numpy as np
arr = np.arange(9).reshape(3,3)
print(arr)

# transpose
arr1 = arr.transpose()
print(arr1)

#
arr1 = arr.T
print(arr1)
```

**Adding and Removing Data**

1. np.concatenate((arr1,arr2))
2. np.vstack((original,new_row)) add new row
3. np.hstack((original,new_col)) add new col
4. np.delete(arr,index_no) delete element by index number

```python
# concatenate using 1d
import numpy as np
arr1 = np.array([10,20,30])
arr2 = np.array([1,2,3])

# concatenate
arr_arr2 = np.concatenate((arr1,arr2))
print(arr_arr2)

# concatenate using 2d axis
arr1 = np.array([[10,20,30],[40,50,60]])
arr2 = np.array([[1,2,3],[4,5,6]])
arr1_arr2 = np.concatenate((arr1,arr2),axis=1)
print(arr1_arr2)
```

```python
# vstack -> vertically join | row wise
import numpy as np
arr1 = np.array([1,2,3])
arr2 = np.array([4,5,6])
result = np.vstack((arr1,arr2))
print(result)
```

```python
# hstack -> horizontally join | column wise
import numpy as np
arr1 = np.array([10,20,30])
arr2 = np.array([40,50,60])
result = np.hstack((arr1,arr2))
print(result)

# same for 2d
```

**Statistical Functions**

1. np.sum(a) -> Sum of all elements

2. np.mean(a) -> Average = Sum of elements / Number of elements

3. np.median(a) -> Middle value after sorting

4. np.min(a) -> Smallest value in array

5. np.max(a) -> Largest value in array

6. np.var(a) mean,difference,square,average | (sum = ddof) -> divide by array length

7. np.std(a) -> Standard Deviation = √Variance -> Measures spread of data

8. np.prod(a) -> Multiplication of all elements

9. np.cumsum(a) -> Cumulative (running) sum

10. np.cumprod(a) -> Cumulative (running) product

11. np.argmin(a) -> Index position of minimum value

12. np.argmax(a) -> Index position of maximum value

13. np.abs(a) -> Converts negative values to positive -> Absolute value (distance from zero)

14. np.unique(a) -> Returns unique values only -> Removes duplicates

15. np.percentile(a, 50) -> Finds percentage-based value -> 50th percentile = Median

16. np.quantile(a, 0.5) -> Similar to percentile -> 0.5 = 50%

17. np.ptp(a) -> Range = Max - Min

18. np.any(a) -> True if at least one value is True

**Handling missing value**

```python
# find missing value
import numpy as np
arr = np.array([10,20,np.nan,40])
print(arr)
# isnan
print(np.isnan(arr))

# update nan by number
new_arr = np.nan_to_num(arr,nan=100)
print(new_arr)
```
