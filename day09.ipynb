# Operations on Rows (loc & iloc)

1. `loc[row, "column"]` – Get a single value using row label and column name.
2. `loc[row]` – Get a complete row.
3. `loc[[row1, row2]]` – Get multiple rows.
4. `loc[[rows], ["col1", "col2"]]` – Get multiple rows and selected columns.
5. `iloc[row_position, column_position]` – Get a single value using index positions.
6. `iloc[row_position]` – Get a complete row using position.
7. `iloc[[row1, row2]]` – Get multiple rows using positions.
8. `iloc[[rows], [cols]]` – Get selected rows and columns using positions.
9. `iloc[[rows], [col]]` – Get a specific column from selected rows.
10. `iloc[[rows], col]` – Get a column as a Series.
11. `iloc[start:end]` – Get rows using slicing.

````python
import pandas as pd

data = {
    "Emp_ID": [201, 202, 203, 204, 205, 206],
    "Name": ["Rahul", "Priya", "Karan", "Sneha", "Aman", "Pooja"],
    "Department": ["IT", "HR", "Finance", "Marketing", "Sales", "IT"],
    "Salary": [55000, 42000, 65000, 48000, 53000, 70000],
    "Experience": [3, 2, 6, 4, 5, 8]
}

df = pd.DataFrame(data)
print(df)
````

# Filter DataFrame

1. Filter using a single condition.
2. Filter using multiple conditions (AND).
3. Filter using multiple conditions (OR).
4. Filter using `isin()`.
5. Filter using `between()`.
6. Filter using string methods (`startswith()`).

````python
# filter
import pandas as pd
data = {
    "Emp_ID": [101, 102, 103, 104, 105, 106],
    "Name": ["Amit", "Riya", "Raj", "Sara", "John", "Neha"],
    "Department": ["IT", "HR", "Finance", "IT", "Sales", "HR"],
    "Salary": [50000, 45000, 60000, 55000, 48000, 52000],
    "Experience": [2, 3, 5, 4, 1, 3]
}
df = pd.DataFrame(data)
print(df)
# filter one greater
print(df[df['Emp_ID'] > 104])

# 106
print(df[df['Emp_ID'] == 106])

# 30000 to 50000
# print(df[(condition one) & (condition 2)])
print(df[(df['Salary']> 30000) & (df['Salary'] < 50000)])
````

````python
import pandas as pd

# Create DataFrame
data = {
    "Emp_ID": [201, 202, 203, 204, 205, 206],
    "Name": ["Rahul", "Priya", "Karan", "Sneha", "Aman", "Pooja"],
    "Department": ["IT", "HR", "Finance", "Marketing", "Sales", "IT"],
    "Salary": [55000, 42000, 65000, 48000, 53000, 70000],
    "Experience": [3, 2, 6, 4, 5, 8]
}

df = pd.DataFrame(data)

print("Original DataFrame")
print(df)

# =====================================================
# 1. Filter by Single Condition
# =====================================================
print("\n1. Salary > 50000")
print(df[df["Salary"] > 50000])

# =====================================================
# 2. Filter by Multiple Conditions (AND)
# =====================================================
print("\n2. Salary > 50000 AND Experience > 4")
print(df[(df["Salary"] > 50000) & (df["Experience"] > 4)])

# =====================================================
# 3. Filter by Multiple Conditions (OR)
# =====================================================
print("\n3. Department = IT OR HR")
print(df[(df["Department"] == "IT") | (df["Department"] == "HR")])

# =====================================================
# 4. Filter Using isin()
# =====================================================
print("\n4. Department in ['IT', 'Finance']")
print(df[df["Department"].isin(["IT", "Finance"])])

# =====================================================
# 5. Filter Using Between()
# =====================================================
print("\n5. Salary Between 45000 and 60000")
print(df[df["Salary"].between(45000, 60000)])

# =====================================================
# 6. Filter Using String Matching
# =====================================================
print("\n6. Names Starting with 'P'")
print(df[df["Name"].str.startswith("P")])
````
